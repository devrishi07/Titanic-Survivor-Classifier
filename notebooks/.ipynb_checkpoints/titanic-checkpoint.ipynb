```python
import matplotlib.pyplot as plt

plt.hist(train["Fare"], bins=50)
plt.xlabel("Fare")
plt.ylabel("Count")
plt.xscale('log')
plt.title("Distribution of Fare")
plt.show()
```

```python
survived = train[train['Survived'] == 1]['Fare']
not_survived = train[train['Survived'] == 0]['Fare']

plt.hist(not_survived, bins=30, alpha=0.6, label="Not Survived")
plt.hist(survived, bins=30, alpha=0.6, label="Survived")
plt.legend()
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Fare Distribution by Survival")
plt.show()
```

Clearly the passengers who paid more had greater chance of survival.
